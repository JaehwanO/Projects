```python
# train_df_sub = train_df[['MSSubClass','LotFrontage']]
from copy import deepcopy
train_df_sub = deepcopy(train_df[['MSSubClass','LotFrontage','SalePrice']])
train_df_sub
```

```python
train_df_sub =train_df_sub.fillna(train_df_sub.mean())
```

```python
X_train, y_train = train_df_sub.to_numpy()[:,:-1],train_df_sub.to_numpy()[:,-1]
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train,y_train)
lr
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#Predicted price
88327 + 60*65 + 65*1260
```

```python
train_df_sub['linear_predictions'] = lr.predict(X_train)
train_df_sub
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(train_df_sub['SalePrice'],train_df_sub['linear_predictions'])
```
